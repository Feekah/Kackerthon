```python
# concatenate train and test data, check 5 random observations
test['subscribed'] = 'test'
data = pd.concat([train, test]).reset_index(drop=True)
data.sample(5)
```

```python
# data shape
data.shape
```

```python
# drop redundant columns
data.drop(['customer_id','contact','day_of_week','emp_var_rate', 'cons_price_idx', 'cons_conf_idx', 'euribor3m', 'nr_employed'], axis=1, inplace=True)
data.head()
```

```python
# label encode categorical features 
cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'month', 'poutcome']

for col in cols:
    data[col] = pd.factorize(data[col])[0]
data.head(10)
```

```python
# get back to train and test set
train_df = data.loc[(data['subscribed'] != 'test')].reset_index(drop = True)
test_df = data.loc[(data['subscribed'] == 'test')].reset_index(drop = True)
```

```python
print(train_df.shape, test_df.shape)
```

```python
X = train_df.drop(['subscribed'], axis=1)
y = train_df['subscribed']
# this solved the bug we had in class, y was an object before 
# which wont work we some models, so we change its type to int
y = y.astype('int') 
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
```

```python
X_train, X_test, y_train,  y_test = train_test_split(X, y, test_size=0.2, random_state=12)
```

```python
print(X_train.shape, y_train.shape)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_test)
predict
```

```python
prob_predict = model.predict_proba(X_test)[:,1]
prob_predict
```

```python
accuracy = accuracy_score(y_test, predict)
accuracy
```

```python
f1 = f1_score(y_test, predict)
f1
```

```python
print('{} accuracy is: {:.4f}'.format(type(model).__name__, accuracy))
print('{} F1 score is: {:.4f}'.format(type(model).__name__, f1))
```

```python
# drop survived column, we added it during concatenation
test_df.drop(['subscribed'], axis=1, inplace=True)
```

```python
# predict on test data
pred = model.predict(test_df)
```
